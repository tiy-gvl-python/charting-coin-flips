##Data Manufacture

```python
import random

def repeater(func, times):
    return [func() for _ in range(times)]
    
def coin_flip():
    return random.randint(0,1)

def new_repeat(func, times, whyle):
    count = 0
    ls = []
    while count <= whyle:
        ls.append([func() for _ in range(times)])
        count += 1
    return ls


vims = repeater(coin_flip, 2**16)

datus = new_repeat(coin_flip, 2**16, 20)
print(3)
```

```python


def split():
    data = []
    for i in range(17):
        data.append(vims[0:2**i])
    return data
data = split()
```

```python

def data_clean(n, datata):
    cleaned_data = []
    for i in datata:
        cleaned_data.append(i.count(n))
    return cleaned_data



test = data_clean(1, datus)
other_test = (data_clean(0, datus))
print(test)
print(other_test)
```

```python
heads = data_clean(0)
tails = data_clean(1)

print(heads)
print(tails)
```

```python
# A test

names = ['2**' + str(i) for i in range(17)]
plt.plot(heads)
plt.plot(tails)
plt.title('A Test')
plt.xticks(range(17), names)
plt.yscale('log')
plt.show()
```

```python
matplotlib.rcParams.update({'font.size': 16, 'font.family': 'serif'})

figure = plt.figure(figsize = (12,8))
plawt = figure.add_subplot(221)
ploot = figure.add_subplot(222)

diff = [i - e for i, e in zip(heads, tails)]
plawt.plot(diff, 'r-', linewidth = 5)
plawt.grid()
plawt.set_title("Differences\n")
plawt.set_xlabel('$x^n$ Times Flipped')
plawt.set_ylabel('Coin Difference')

ratio = [i / e for i, e in zip(heads, tails)]

ploot.plot(ratio, color = 'k', marker = 'o', markerfacecolor = 'r', markersize = 5)
ploot.grid()
ploot.set_title('Ratios\n')
ploot.set_xlabel('$x^n$ Times Flipped')
ploot.set_ylabel('Coin Ratio')
plt.show()
```
