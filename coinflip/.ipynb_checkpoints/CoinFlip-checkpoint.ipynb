```python
import random
import matplotlib.pyplot as plt

def coinflip(numberofflips):
    coin = ["Heads", "Tails"]
    flips = [random.choice(coin) for item in range(numberofflips)]
    return flips

def ratio(flips):
    head = flips.count("Heads")
    tail = flips.count("Tails")
    return (head, tail)
def flipping(grow):
    heads_number = []
    tails_number = []
    for item in grow:
        heads_number.append(ratio(coinflip(item))[0])
        tails_number.append(ratio(coinflip(item))[1])
    return (sorted(tails_number), sorted(heads_number))
def means():
    number_of_heads = sum(heads_number)
    number_of_tails = sum(tails_number)
    total = number_of_tails + number_of_heads
    print(total)
    rati = number_of_heads/total
    return rati
```

```python
grow = [2**0, 2**1, 2**2, 2**4, 2**8, 2**16]
tup = flipping(grow)
x1 = grow
y1 = tup[1]
z1 = tup[0]
plt.plot(y1)
plt.plot(z1)
plt.xticks(range(6), x1)
plt.xlabel("Growing By")
plt.ylabel("Frequency")
plt.title("Coin")
plt.legend(["Tails Number", "Heads Number"], loc='upper left')
plt.show()
```

##The graph above appears to have 1 line.  This is because the the difference between the number of times tails is flipped and the number of times heads is flipped is so small that they never deviate on the graph enough to see.  
##This occurs because they have even chances
###The graph is also on a very large scale making it hard to view small deviations

```python
grow = [2**0, 2**1, 2**2, 2**4, 2**8]
tup = flipping(grow)
x1 = grow
y1 = tup[1]
z1 = tup[0]
plt.plot(y1)
plt.plot(z1)
plt.xticks(range(5), x1)
plt.xlabel("Growing By")
plt.ylabel("Frequency")
plt.title("Coin")
plt.legend(["Tails Number", "Heads Number"], loc='upper left')
plt.show()
```

###Here we can see the small deviations in the data as tails outnumbered heads for this particular set of data

```python
grow = [2**0, 2**1, 2**2, 2**4]
tup = flipping(grow)
x1 = grow
y1 = tup[1]
z1 = tup[0]
plt.plot(y1)
plt.plot(z1)
plt.xticks(range(4), x1)
plt.xlabel("Growing By")
plt.ylabel("Frequency")
plt.title("Coin")
plt.legend(["Tails Number", "Heads Number"], loc='upper left')
plt.show()
```
