```python
import random
import statistics as stats
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
def coin_flips(number=1):
    for i in range(number):
        yield random.randint(0, 1)

def exponential_coin_flips(exponent):
    return [sum(list(coin_flips(2**exponent))[0:2**exp]) for exp in range(exponent + 1)]

def coin_flip_diff(heads_count):
    return [2*heads - 2**exp for exp, heads in enumerate(heads_count)]

def coin_flip_ratio(heads_count):
    ratio_list = []
    for exp, heads in enumerate(heads_count):
        if 2**exp == heads:
            ratio_list.append(float("inf"))
            continue
        ratio_list.append(heads/(2**exp - heads))
    return ratio_list
```

```python
heads_count = exponential_coin_flips(16)
diff = coin_flip_diff(heads_count)
ratio = coin_flip_ratio(heads_count)
```

```python
plt.plot([2**i for i in range(17)], diff)
plt.title("Difference Between Heads and Tails")
plt.ylabel("Difference")
plt.xlabel("Number of Trials")
plt.show()
```

```python
plt.plot([2**i for i in range(17)], ratio)
plt.title("Ratio Between Heads and Tails")
plt.ylabel("Ratio")
plt.xlabel("Number of Trials")
plt.show()
```

```python
plt.scatter([2**i for i in range(17)], diff)
plt.title("Difference Between Heads and Tails")
plt.ylabel("Difference")
plt.xlabel("Number of Trials")
plt.xscale("log")
plt.show()
```

```python
plt.scatter([2**i for i in range(17)], ratio)
plt.title("Ratio Between Heads and Tails")
plt.ylabel("Ratio")
plt.xlabel("Number of Trials")
plt.xscale("log")
plt.show()
```

```python
coin_flips_trial_2 = [exponential_coin_flips(16) for i in range(20)]
```

```python
ratio_list_trial_2 = [coin_flip_ratio(heads_list) for heads_list in coin_flips_trial_2]
mean_ratios = [stats.mean([trial[exp] for trial in ratio_list_trial_2]) for exp in range(17)]
stdev_ratios = [stats.stdev([trial[exp] for trial in ratio_list_trial_2]) for exp in range(17)]
```

```python
plt.scatter([2**i for i in range(17)], stdev_ratios)
plt.title("Standard Devation of Ratios Between Heads and Tails")
plt.ylabel("Ratio")
plt.xlabel("Number of Trials")
plt.xscale("log")
plt.show()
```

```python
coin_flips_100 = [list(coin_flips(100)) for i in range(100000)]
coin_flips_1000 = [list(coin_flips(1000)) for i in range(100000)]
coin_flips_mean_100 = list(map(stats.mean, coin_flips_100))
coin_flips_mean_1000 = list(map(stats.mean, coin_flips_1000))
```
